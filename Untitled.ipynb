```python
import tensorflow as tf
sess_config = tf.ConfigProto()
sess_config.gpu_options.allow_growth = True
seq_len = tf.placeholder(tf.int32, shape=[None])
```

```python
max_len = tf.reduce_max(seq_len)
rang = tf.range(max_len,dtype=tf.int32)
func = lambda x: tf.to_int32(tf.less(rang, x))
mask = tf.map_fn(func, seq_len, dtype=tf.int32)
```

```python
sess = tf.Session(config=sess_config)
sess.run(tf.global_variables_initializer())
b = sess.run(mask, feed_dict={seq_len:[5,3,2,6]})
```

```python
import itertools
two_dim = tf.placeholder(tf.float32, shape=[4, None])
c_comb = tf.constant(list(itertools.combinations(range(4), 2)), dtype=tf.int32)
c_gather = tf.gather(two_dim, c_comb)
sess.run(c_gather, feed_dict={two_dim: np.arange(12).reshape(4,3)})
```

```python
two_dim = tf.placeholder(tf.int32, shape=[4, None])
a = tf.range(4, dtype=tf.int32)
arg_idx = tf.cast(tf.argmin(two_dim, axis=1), tf.int32)
cc = tf.stack([a, arg_idx], axis=1)
sess.run(cc, feed_dict={two_dim: np.array([[1,2,3,4],[2,1,4,3],[4,1,2,3],[1,4,2,3]])})
```
